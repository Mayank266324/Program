```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Sin Wave")
plt.subplot(1,2,2)
plt.plot(x,z)
plt.title("Cos Wave")
plt.show()
```

```python
# Generate example data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
# Create a figure with two subplots (2 rows, 1 column)
fig, axs = plt.subplots(2, 1,figsize=(6,8))
# Plot the sine wave in the first subplot
axs[0].plot(x, y1, color='blue')
axs[0].set_title('Sine Wave')
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')
# Plot the cosine wave in the second subplot
axs[1].plot(x, y2, color='red')
axs[1].set_title('Cosine Wave')
axs[1].set_xlabel('X')
axs[1].set_ylabel('Y')
# Adjust layout to prevent overlap
plt.tight_layout()
# Show the plot
plt.show()
```
