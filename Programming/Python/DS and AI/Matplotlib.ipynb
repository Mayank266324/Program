```python
a=(5)
b=(5)

x=(5,15,25)
y=(10,50,12)

c=(5,15,25)
d=(10,50,90)


plt.subplot(2,3,1) # 2= no of rows, 3 no of column, 1 position
plt.plot(a,b,"D")
plt.subplot(2,3,2)
plt.plot(x,y,marker="P",linestyle="dashed")
plt.subplot(2,3,3)
plt.plot(x,y,"o",color="green")
plt.subplot(2,3,4)
plt.barh(x,y,height=0.5,color="red")
plt.subplot(2,3,5)
plt.plot(c,d,"D")
plt.subplot(2,3,6)
plt.bar(x,y,width=4)
plt.show()
```

```python
a=(5)
b=(5)

x=(5,15,25)
y=(10,50,12)

c=(5,15,25)
d=(10,50,90)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Sub-Plot")
plt.subplot(2,3,1) # 2= no of rows, 3 no of column, 1 position
plt.plot(a,b,"D")
plt.subplot(2,3,2)
plt.plot(x,y,marker="P",linestyle="dashed")
plt.subplot(2,3,3)
plt.plot(x,y,"o",color="green")
plt.subplot(2,3,4)
plt.barh(x,y,height=0.5,color="red")
plt.subplot(2,3,5)
plt.plot(c,d,"D")
plt.subplot(2,3,6)
plt.bar(x,y,width=4)
plt.tight_layout()
plt.show()
```
