```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
subject=['science','english','math','evs']
perctange=[78,89,60,54]
plt.pie(perctange,labels=subject,autopct='%1.1f%%')
plt.show()
```

```python
#hexbin plot
x=np.random.randn(1000)
y=np.random.randn(1000)
plt.hexbin(x,y,gridsize=30,cmap='Reds')
plt.colorbar(label='count')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Hexagonal bin plot")
plt.show()
```
